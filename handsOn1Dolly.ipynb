```python
import pandas as pd
import matplotlib.pyplot as pit
url='/content/Test.csv'
df = pd.read_csv(url)
df
```

```python
mean_value = df['Item_MRP'].mean()
print(f"mean value:{mean_value}")
```

```python
median_value = df['Item_MRP'].median()
print(f"median value:{median_value}")
```

```python
mode_value = df['Item_MRP'].mode()
print(f"mode value:{mode_value}")
```

```python
var_value = df['Item_MRP'].var()
print(f"variance value:{var_value}")
```

```python
sd_value = df['Item_MRP'].std()
print(f"standard deviation value:{sd_value}")
```

```python
skew_value = df['Item_MRP'].skew()
print(f"skewness value:{skew_value}")
```

```python
Kur_value = df['Item_MRP'].kurtosis()
print(f"Kurtosis :{Kur_value}")
```

```python
df['Item_MRP'].value_counts().plot(kind='bar',title='Exam Score Count')
pit.xlabel('Item_MRP')
pit.ylabel('Count')
pit.show()
```
